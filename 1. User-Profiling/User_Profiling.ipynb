```python
from pynput import keyboard

err = 0
charac = 0
t0 = time.time()
t1 = 0.0
err = 0
char = 0

def on_press(key):
    #global t0
    global err
    global char   
    
    
    try:
        print('alphanumeric key {0} pressed'.format(
            key.char))
        if (key == keyboard.Key.backspace or key == keyboard.Key.delete):
            err+=1
            #print("***********************")
        else:
            char+=1
            #print("***********************")
    except AttributeError:
        if (key == keyboard.Key.backspace or key == keyboard.Key.delete):
            err+=1
            #print("***********************")
        print('special key {0} pressed'.format(
            key))

def on_release(key):
    global t1
    print('{0} released'.format(
        key))
    if key == keyboard.Key.esc:
        # Stop listener
        t1 = time.time()
        return False

# Collect events until released
with keyboard.Listener(
        on_press=on_press,
        on_release=on_release) as listener:
    listener.join()


listener = keyboard.Listener(
    on_press=on_press,
    on_release=on_release)
listener.start()

#print(t1)
#print(t0)
#print(err)
#print(char)
minutes = (t1-t0)/60.0
err_rate = err/minutes
char_rate = char/minutes
print(minutes)
print(char_rate)
print(err_rate)


```
